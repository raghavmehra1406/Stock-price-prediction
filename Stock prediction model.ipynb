```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the stock ticker (In this case im taking Tata Motors)
ticker = 'TATAMOTORS.BO'

# Download the stock data for the past 5 years
data = yf.download(ticker, start='2018-01-01', end='2023-01-01')

# Display the first few rows of data
print(data.head())
```

```python
# Plot the closing price
plt.figure(figsize=(10,6))
plt.plot(data['Close'])
plt.title(f'{ticker} Stock Price')
plt.xlabel('Date')
plt.ylabel('Closing Price (INR)')
plt.show()
```

```python
# Calculate the 50-day Simple Moving Average
data['SMA_50'] = data['Close'].rolling(window=50).mean()

# Plot the closing price and the 50-day SMA
plt.figure(figsize=(10,6))
plt.plot(data['Close'], label='Closing Price')
plt.plot(data['SMA_50'], label='50-day SMA', color='red')
plt.title(f'{ticker} Stock Price with 50-day SMA')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.show()
```

```python
# Ploting the volume traded
plt.figure(figsize=(10,6))
plt.plot(data['Volume'])
plt.title(f'{ticker} Stock Volume Traded')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()
```

```python
'''From the analysis, I found that Tata Motors' stock price has had some ups and downs over the last few years. The 50-day moving average helped show the overall trend making it easier to spot if the stock is  going up or down generally speaking .

I also noticed that when there was a lot of trading volume, the price often moved more, which could mean something big was happening with the company or the market.

In short, the stock price shows some clear trends, and the moving average gives a good idea of whether it's on the rise or fall in the long term.'''
```
